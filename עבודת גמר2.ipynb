```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns;  sns.set(style="ticks", color_codes=True)
```

```python
# שני גרפים אחד ע"פ הדירוג ואחד ע"פ דירוג הגרסא
appleStore= pd.read_csv("Data//AppleStore.csv")
plt.hist(appleStore['user_rating'],facecolor='Blue',bins=30)
plt.hist(appleStore['user_rating_ver'],facecolor='red',alpha=0.3,bins=30)
plt.legend(['total rating', 'last virsion rating'], loc=2)
plt.show
```

```python
#פאי ע"פ כמות הרייטינג 
plt.rcdefaults()
n=appleStore.groupby(['user_rating']).count()

values = n['price']
#explode = [0, 0, 0.2, 0, 0]
labels = ['0','1','1.5','2','2.5','3','3.5','4','4.5','5']
plt.pie(values, labels=labels)
plt.title('rating')
plt.show()
```

```python
#מספר השורות על פי כל קטגוריה
p=appleStore.groupby(['prime_genre']).count()
p['price'].plot.bar()
plt.show()
```

```python
#גרף שמציג את הדירוג הממוצע של הדירוג גרסא ביחס לדירוג הכללי
m=appleStore.groupby(['user_rating']).mean()
plt.plot(m['user_rating_ver'].tolist())
plt.ylabel('user rating version')

#my_xticks = ['0','Arnold','Mavis','Matt']
#plt.xticks(x, my_xticks)

plt.show()
```
